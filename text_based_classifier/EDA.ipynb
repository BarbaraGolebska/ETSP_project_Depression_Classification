# Explore transcript structure

Take a look at the first transcript to get the idea of the structure.

```python
transcript = pd.read_csv('../daic_data/EDAIC_text/transcripts/300_Transcript.csv')
transcript
```

We can see that the transcript is auto-generated from the participant's speech. There are a couple of problems with that. Additionally, speech outside of the formal interview is included in the transcript as well (since it is included in the audio). The "Start_Time" in the last row is also labeled incorrectly, so the data is not perfect. Let us see another file.

```python
pd.read_csv('../daic_data/EDAIC_text/transcripts/301_Transcript.csv') 
```

It looks like the issue with the last row at "Start_Time" is a repeating problem. It appears it duplicates the time from the first row. This issue would most likely be irrelevant for our task, however.

# Explore labels directory

## Explore detailed labels

```python
detailed_lables = pd.read_csv('../daic_data/EDAIC_text/labels/detailed_lables.csv')
detailed_lables
```

```python
detailed_lables.info()
```

Focus on the target for our task:

```python
detailed_lables['Depression_label'].value_counts(normalize=True)
```

Is is evident that the classes are unbalanced. Let us explore the distribution for each split.

```python
detailed_lables.groupby('split')['Depression_label'].value_counts(normalize=True)
```

The distribution for labels is roughly the same for train and dev, and there is a slightly larger proportion of 'depressed' in test.
